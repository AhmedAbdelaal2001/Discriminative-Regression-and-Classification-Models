```python
from utility import load_dataset, add_polynomialTerms, add_sine_poly, plot
from LinearModel import LinearRegression
```

```python
X_train, y_train = load_dataset("training_set.csv")
X_valid, y_valid = load_dataset("dev_set.csv")
X_test, y_test = load_dataset("test_set.csv")
```

```python
plot(X_train, y_train, "Training Set")
```

```python
k_arr = [3, 5, 10, 20]
linReg = LinearRegression()
X_train_temp1 = X_train
```

```python
for k in k_arr:
    print(f"Polynomial Degree = {k}.")
    X_train_temp1 = add_polynomialTerms(X_train_temp1, k)
    linReg.fit(X_train_temp1, y_train)
    plot(X_train_temp1, y_train, "Training Set", linReg, k)
    plot(X_valid, y_valid, "Validation Set", linReg, k)
```

```python
k_arr = [1, 2, 3, 5, 10, 20]
linReg = LinearRegression()
```

```python
linReg.fit(add_sine_poly(X_train, k = 0), y_train)
plot(X_train, y_train, "Training Set", linReg, 0, sine = True)
plot(X_train, y_train, "Validation Set", linReg, 0, sine = True)

for k in k_arr:
    print(f"Degree of the polynomial = {k}.")
    X_train = add_sine_poly(X_train, k)
    linReg.fit(X_train, y_train)
    plot(X_train, y_train, "Training set", linReg, k, sine = True)
    plot(X_train, y_train, "Validation Set", linReg, k, sine = True)    
```
