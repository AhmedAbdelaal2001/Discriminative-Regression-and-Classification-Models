```python
from model import LogisticRegression
from utility import load_dataset, plot, accuracy, save_parameters
```

```python
X_train1, y_train1 = load_dataset("training_set1.csv", addIntercept = True)
X_train2, y_train2 = load_dataset("training_set2.csv", addIntercept = True)

X_valid1, y_valid1 = load_dataset("dev_set1.csv")
X_valid2, y_valid2 = load_dataset("dev_set2.csv")
```

```python
plot(X_train1, y_train1, x_min = -4, y_min = -2, x_max = 6, y_max = 80, model = None)
```

```python
plot(X_train2, y_train2, x_min = -3, y_min = -3, x_max = 6, y_max = 6, model = None)
```

```python
logReg1 = LogisticRegression()
logReg2 = LogisticRegression()

logReg1.fit(X_train1, y_train1)
print("--------------------------------------------------------------------------")
logReg2.fit(X_train2, y_train2)
```

```python
plot(X_valid1, y_valid1, x_min = -2, x_max = 5, y_min = -2, y_max = 40, model = logReg1)
```

```python
plot(X_valid2, y_valid2, x_min = -2, x_max = 5, y_min = -2, y_max = 5, model = logReg2)
```

```python
y_pred1_valid = logReg1.predict_binary(X_valid1, addIntercept = True)
y_pred1_train = logReg1.predict_binary(X_train1)

y_pred2_valid = logReg2.predict_binary(X_valid2, addIntercept = True)
y_pred2_train = logReg2.predict_binary(X_train2)
```

```python
print(f"Accuracy of model 1 on its corresponding validation set = {accuracy(y_pred1_valid, y_valid1) * 100}%")
print(f"Accuracy of model 1 on its training set = {accuracy(y_pred1_train, y_train1) * 100}%")
print(f"Accuracy of model 2 on its corresponding validation set = {accuracy(y_pred2_valid, y_valid2) * 100}%")
print(f"Accuracy of model 2 on its training set = {accuracy(y_pred2_train, y_train2) * 100}%")
```

```python
save_parameters(logReg1, logReg2)
```
