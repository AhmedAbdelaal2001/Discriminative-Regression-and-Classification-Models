```python
import numpy as np
from utility import add_intercept, load_dataset_posonly, plot, accuracy, predictPosonly
from model import LogisticRegression
```

```python
X_train, t_train = load_dataset_posonly("training_set3.csv", addIntercept = True, y_firstColumn = True)
X_valid, t_valid = load_dataset_posonly("dev_set3.csv", y_firstColumn = True)
X_test, t_test = load_dataset_posonly("posonly_test_set.csv", y_firstColumn = True)
```

```python
plot(X_train, t_train, x_min = -8, y_min = -8, x_max = 5, y_max = 8)
```

```python
logReg = LogisticRegression()
logReg.fit(X_train, t_train)
```

```python
plot(X_valid, t_valid, x_min = -8, y_min = -8, x_max = 5, y_max = 8, model = logReg)
```

```python
plot(X_test, t_test, x_min = -8, y_min = -8, x_max = 5, y_max = 8, model = logReg)
```

```python
t_pred_train = logReg.predict_binary(X_train)
t_pred_valid = logReg.predict_binary(X_valid, addIntercept = True)
t_pred_test = logReg.predict_binary(X_test, addIntercept = True)
```

```python
print(f"Accuracy of the model on the training set = {accuracy(t_pred_train, t_train) * 100}%")
print(f"Accuracy of the model on the validation set = {accuracy(t_pred_valid, t_valid) * 100}%")
print(f"Accuracy of the model on the test set = {accuracy(t_pred_test, t_test) * 100}%")
```
